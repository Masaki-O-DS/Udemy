#### ①　以下の二次方程式を二分法で解き、その解の推移をグラフでプロット　x**2-3=0

```python
import sympy
import math
x = sympy.Symbol('x')
result = sympy.solve(x**2-3)[1]**2
print(math.sqrt(result))
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x = np.linspace(0, 4, 5)
y = x ** 2 - 3

x1 = -10
x2 = 10
x3 = (x1+x2) / 2
e = 0.01

y1 = x1**2-3
y2 = x2**2-3
y3 = x3**2-3

x_list = []
y_list = []

while abs(x2-x1) > e:
    if y2 * y3 < 0:
        x1 = x3
        x3 = (x1+x2) / 2
        y1 = x1 ** 2-3
        y3 = x3 ** 2-3
    elif y1*y3 < 0:
        x2 = x3
        x3 = (x1+x2) / 2
        y2 = x2**2 - 3
        y3 = x3**2 - 3
        
    x_list.append(x3)
    y_list.append(y3)
    
plt.xlabel('count')
plt.ylabel('x3')
plt.title('x3 transition')
plt.grid()
plt.plot(x_list)
print(f'解は{x3}')
```

#### ② １の二次方程式をニュートン法で解き、その解の推移をグラフでプロット　x**2-3

```python
x = np.array([6.0])
e = 0.1 # 近似値
i = 0   #インデックス番号

while True:
    result = x[i] - (x[i]**2-3)/(2*x[i])
    x = np.append(x, result)
    if np.abs(x[i+1] - x[i]) < e:
        break
    i += 1
plt.plot(x)
plt.grid()
plt.xlabel('count')
plt.ylabel('x')
print(f'解は{x[-1]}')
```

### ③　以下の五次方程式を二分法で解き、その解の推移をグラフでプロット　2x**5-3x**4-6x-99=0

```python
x = np.linspace(0, 4, 5)
y = 2*x**5-3*x**4-6*x-99

x1 = -10
x2 = 10
x3 = (x1+x2) / 2
e = 0.01
# for i in range(1,3):
#     y[i] = 2*x[i]**5-3*x[i]**4-6*x[i]-99
y1 = 2*x1**5-3*x1**4-6*x1-99
y2 = 2*x2**5-3*x2**4-6*x2-99
y3 = 2*x3**5-3*x3**4-6*x3-99
x_list = [-10, 10, (x1+x2)/2]
y_list = []
while abs(x2-x1) > e:
    if y2 * y3 < 0:
        x1 = x3
        x3 = (x1+x2) / 2
        y1 = 2*x1**5-3*x1**4-6*x1-99
        y3 = 2*x3**5-3*x3**4-6*x3-99
    elif y1*y3 < 0:
        x2 = x3
        x3 = (x1+x2) / 2
        y2 = 2*x2**5-3*x2**4-6*x2-99
        y3 = 2*x3**5-3*x3**4-6*x3-99
        
    x_list.append(x3)
    y_list.append(y3)
    
plt.xlabel('count')
plt.ylabel('x3')
plt.title('x3 transition')
plt.grid()
plt.plot(x_list)
print(f'解は{x3}')
```

#### ④ ３の五次方程式をニュートン法で解き、その解の推移をグラフでプロット

```python
x = np.array([6.0])
e = 0.1 # 近似値
i = 0   #インデックス番号

while True:
    result = x[i] - (2*x[i]**5-3*x[i]**4-6*x[i]-99)/(10*x[i]**4-12*x[i]**3-6)
    x = np.append(x, result)
    if np.abs(x[i+1] - x[i]) < e:
        break
    i += 1
plt.plot(x)
plt.grid()
plt.xlabel('count')
plt.ylabel('x')
print(f'解は{x[-1]}')
```

#### ⑤　f(x)=sinxからf'(x)を中心差分近似を使い、両方のグラフをプロット

####  確認用　sinxのグラフ

```python
x = np.linspace(-10, 10, 100)
y = np.sin(x)
plt.grid()
plt.plot(x, y)
```

####  前進差分近似  確認用

```python
# 前進差分近似
h = 0.01
xd = [math.pi*i/100 for i in range(-200,201)]
yd = [math.sin(x) for x in xd]
yd_diff = [(math.sin(x+h)-math.sin(x))/h for x in xd ]

plt.plot(xd, yd_diff, label='cosx')
plt.plot(xd,yd, label='sinx')
plt.grid()
plt.legend(loc=4)
```

#### 後退差分近似  確認用

```python
h = 0.01
xd = [math.pi*i/100 for i in range(-200,201)]
yd = [math.sin(x) for x in xd]
yd_diff = [(math.sin(x)-math.sin(x-h))/h for x in xd ]

plt.plot(xd, yd_diff, label='cosx')
plt.plot(xd,yd, label='sinx')
plt.grid()
plt.legend(loc=4)
plt.show()
```

### 中心差分近似

```python
h = 0.01
xd = [math.pi*i/100 for i in range(-200,201)]
yd = [math.sin(x) for x in xd]
yd_center_diff = [(math.sin(x+h)-math.sin(x-h))/(2*h) for x in xd ]  # 中心差分近似
plt.plot(xd, yd_center_diff, label='cosx')
plt.plot(xd,yd, label='sinx')
plt.grid()
plt.legend(loc=4)
```

### ⑥f(x)=x3-6x2+x+10からf'(x)を中心差分近似を使い、両方のグラフをプロット

#### 確認用

```python
xd = [i for i in range(-200,201)]
y1 = [x**3-6*x**2+x+10 for x in xd]
y = [3*x**2-12*x+1 for x in xd]
plt.plot(xd, y)
plt.plot(xd, y1)
plt.grid()
```

```python
h = 0.01
xd = [i for i in range(-200,201)]
yd = [x**3-6*x**2+x+10 for x in xd]
yd_center_diff = [(((x+h)**3-6*(x+h)**2+(x+h)+10)-((x-h)**3-6*(x-h)**2+(x-h)+10))/(2*h) for x in xd ]  # 中心差分近似
plt.plot(xd, yd_center_diff, label='3*x**2-12*x+1')
plt.plot(xd,yd, label='x**3-6*x**2+x+10')
plt.grid()
plt.legend(loc=4)
```
